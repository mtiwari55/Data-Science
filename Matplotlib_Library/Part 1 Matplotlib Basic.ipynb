# Basic information for Matplotlib

```python
import matplotlib.pyplot as plt

#pyplot is class of matplotlib which helps to create graph
```

```python
# linear values

x = [1,2,3,4,5,6,7,8]
y = [1,2,3,4,5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
# add title
plt.plot(x,y)
plt.title("lenear plot")
plt.show()
```

```python
# add x -axis lable and y axis label

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.title("lenear plot")
plt.show()
```

```python
# randam value

x = [1,2,3,4,5,6,7,8]
y = [1,4,3,7,1,5,8,2]

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.title("random plot")
plt.show()
```
