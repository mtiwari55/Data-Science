# Read and write excel 

```python
import pandas as pd
import os
getpath = os.getcwd()

def EnFun(ent):
    if ent == "":
        return "Normal"
    
    return ent

def TemFun(temp):
    if temp == "":
        return 0
    
    return temp

exce = pd.read_excel("C:\\Users\\mantiwar\\Desktop\\Pythan-Prac\\weather1.xlsx" , converters = {
    "Event":EnFun ,
    "Temp" : TemFun
})

exce

# type(exce.Date[0])
```

```python
exce
```

```python
# export records in new excel files
# Noraml exports

exce.to_excel("exportExcel.xlsx")
```

```python
#  i want start from 2 row and 3 column

exce.to_excel("exportExcel.xlsx" , startrow=2 , startcol=3 ,index=False)
```

```python
# add sheet name


exce.to_excel("exportExcel.xlsx" , startrow=2 , startcol=3 ,index=False ,sheet_name="wheat")
```
