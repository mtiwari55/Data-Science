# Pandas Stack and unstack

```python

# import pandas module 
import pandas as pd 
  
# making dataframe 
df = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 
  
# it was print the first 5-rows 
df 
```

```python
df.dropna(thresh =1 ,inplace =True)

df
```

```python
df.fillna({
    "College":"Local Collage"
} )
```

```python
# same collage bellong

coll = df.groupby("College")

coll
```

```python
for collage ,Detail in coll:
    print(collage)
    print(Detail)
```

```python
df_stacked = df.stack()
```
