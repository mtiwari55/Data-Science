# Combining DataFrames

```python
import pandas as pd

column = ["Name" , "Address" ,"Gender","Phone"]
column
```

```python
stud1 = pd.DataFrame([
           ["Manish","Mumbai","M" ,12345],
           ["Deepak" ,"Goa" ,"M" ,123456]
         ],columns=column)
stud1
```

```python
stud2 = pd.DataFrame( 
                     dict(Name=['peter', 'julie'],
                           Address=[33, 44], 
                           Gender=['M', 'F'],
                          Phone=[123456, 1234567])
                   )

stud2
```

```python
# concat 

stud = pd.concat([stud1,stud2])
stud
```

```python
# merge

stud2 = pd.DataFrame(dict(
                        Name =["Deepak","Priya" ,"Gita","Rita"],
                        Address=["Mumbai","Chembure" ,"Navi Mumbai",""],
                        Gender = ["M","F","F","F"],
                        Phone = [123,456,789,4123]
                  ))
stud2
```

```python
stud 
```

```python
join = pd.merge(stud,stud2 ,on="Name" ,how="outer")
join
```
