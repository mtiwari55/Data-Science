```python
# read copy csv file

csvCopy = pd.read_csv(getpath+"//Copy.csv")
```

```python
csvCopy.head(20)
```

```python
# find length and apply as period length

length = len(csvCopy)

date = pd.date_range("01-01-2019" ,periods=length ,freq="D")
```

```python
date
length
```

```python
csvCopy.set_index(date ,inplace =True)
```
