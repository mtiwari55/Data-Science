```python
# custom holiday list

from pandas.tseries.holiday import *
from pandas.tseries.offsets import CustomBusinessDay

class myBirthdayCalendar(AbstractHolidayCalendar):
   rules = [
     Holiday('Manish jaintiy', month=10, day=4, observance=sunday_to_monday),
   ]

BD = CustomBusinessDay(calendar=myBirthdayCalendar())  
```

```python
dateRange = pd.date_range('01-01-2019' ,periods=lenght, freq=BD)
```

```python
dateRange
```

```python
recodss =cvs.set_index(dateRange)
```
